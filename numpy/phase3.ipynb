```python
sales_data = np.array([
    [1,150000,180000,220000,250000], #paradise biryani
    [2,120000,140000,160000,190000], #beijing bites
    [3,200000,230000,260000,300000], #pizza hub
    [4,180000,210000,240000,270000],#burger point
    [5,160000,185000,205000,230000] #chai point
])
print("=== Zomato sale analysis ===")
print("\n sales data dhape",sales_data.shape)
print("\nsample data for first 3 resturant",sales_data[0:3])
```

```python
#total sale per year-sum of all resturants combined in each year
print(np.sum(sales_data ,axis=0))
# yearly total
yearly_total=np.sum(sales_data[:, 1:] , axis=0) 
print(yearly_total)
#total revenue of each resturants in all 4 years combined
res_total = np.sum(sales_data[:, 1:],axis=1) # adds all terms of each row
print(res_total)

#minimum sales per resturant
min_res = np.min(sales_data[:, 1:],axis =1)
print(min_res)

#resturant with minimum sale in each year
min_in_year = np.min(sales_data[: ,1:],axis=0)
print(min_in_year)

#average sale per resturant in given years
avg_res = np.average(sales_data[:, 1:], axis=1)
print(avg_res)
```

```python
cumsum = np.cumsum(sales_data[:,1:],axis =1) #revenue of this year+previous years
print(cumsum)

plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("average cumalitve sales across all res")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```

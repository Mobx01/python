## Numpy array and basics

```python
import numpy as np 
```

### creating array from list

```python
arr_1d = np.array([1,2,3,4,5])
print("1D array : " ,arr_1d)

arr_2d = np.array([[1,2,3] ,[4,5,6]])
print("2d array :" ,arr_2d)
```

### List vs numpy array

```python
py_list = [1,2,3]
print("python list multiplication",py_list *2)

np_array = np.array([1,2,3]) #element wise multiplication
print("numpy array multiplication",np_array * 2) # multiplication with array

import time
start = time.time();
py_list = [i*2 for i in range(1000000)]
print("\n List operation time ",time.time()-start)

start =time.time()
np_array = np.arange(1000000) *2
print("\n np array operation time ",time.time()-start)
```

### creating array from scratch

```python
zeros = np.zeros((3,4))
print("\nzeros matrix",zeros)

ones = np.ones((2,3))
print("\nones matrix :",ones)

full = np.full((3,5),10)
print("Matrix with same value:" ,full)

random = np.random.random((4,3))
print("\nMatrix with random value:" ,random)

sequence = np.arange(0,10,2)
print("\n sequnce array :" ,sequence)
```

### vector , Matrix and Tensor

```python
vector = np.array([1,2,3])
print("\n Vetor :",vector)

matrix = np.array([[1,2,3],
                   [3,4,5]])
print("\n Matrix :",matrix)

tensor = np.array([[[1,3,4],[1,2,3]],
                    [[4,7,8],[3,9,1]]
                    ])
print("\n Tensor :",tensor)
```

### array properties


```python
arr = np.array([[1,2,3],
                [4,5,6]
                ])
print("shape:" ,arr.shape)
print("dimension :",arr.ndim)
print("size :", arr.size)
print("data type :", arr.dtype)
```

### array reshaping

```python
arr = np.arange(12)
print("orignal array :" ,arr)

reshaped = arr.reshape((3,4))
print("reshaped array :" ,reshaped)

flatten = reshaped.flatten()
print("flattened array :" ,flatten)

# ravel return view instead of copy
raveled = reshaped.ravel()
print("raveled array :" ,raveled)

#transpose
transpose = reshaped.T
print("Transposed array :" ,transpose)
```
